```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
food_df = pd.read_csv(r"C:\Users\91903\Documents\Winter PEP\1612\indian_food.csv")
food_df.head()
```

Count of Vegetarian vs Non-Vegetarian Dishes

```python
food_df['diet'].value_counts().plot(
    kind='bar',
    title='Vegetarian vs Non-Vegetarian Dishes',
    xlabel='Diet Type',
    ylabel='Count'
)
plt.show()
```

Count of Unique Dishes

```python
unique_dishes = food_df['name'].nunique()
print("Total unique dishes:", unique_dishes)
```

```python
food_df.groupby('course')['name'].nunique().plot(kind='bar', figsize=(10,5))
plt.title("Unique Dishes per Category")
plt.ylabel("Number of Unique Dishes")
plt.xlabel("Category")
plt.show()
```

```python
food_df.groupby('region')['name'].nunique().plot(
    kind='bar', figsize=(8,5)
)
plt.title("Unique Dishes by Region")
plt.xlabel("Region")
plt.ylabel("Number of Unique Dishes")
plt.show()
```

```python
food_df.groupby('state')['name'].nunique().sort_values(
    ascending=False
).head(10).plot(kind='bar', figsize=(10,5))

plt.title("Top 10 States by Unique Dishes")
plt.xlabel("State")
plt.ylabel("Number of Unique Dishes")
plt.show()
```

Count of Unique Flavored Dishes

```python
unique_flavors = food_df['flavor_profile'].nunique()
print("Total unique flavor profiles:", unique_flavors)
print("Unique flavor profiles:", food_df['flavor_profile'].unique())
```

```python
food_df['flavor_profile'].value_counts().plot(
    kind='bar',
    title='Flavor Profile Distribution',
    xlabel='Flavor Profile',
    ylabel='Count'
)
plt.show()
```
